# Read a file
Before pulling up the file, I want to choose coltype because FIPS CODE is numerical code, but many programs recoginized as numerical code, so it deletes leading zero automatically.

```python
# Preselect column type as object
coltype = {'SUMLEV':object,
          'STATE':object,
           'COUNTY':object,
          'PLACE':object,
          'COUSUB':object,
          'CONCIT':object,
          'PRIMGEO_FLAG':object}
```

```python
pop_2018 = pd.read_csv("sub-est2018_all.csv",encoding="ISO-8859-1", dtype= coltype)
```

```python
pop_2018.head()
```

```python
# First, I want to find out how many state FIPS code. I know that it should be 51 including DC.
pop_2018['STATE'].nunique()
```

# There are 51 unique state codes, which is right.

```python
pop_2018.groupby(['STATE', 'STNAME','COUNTY','PLACE']).agg(['count'])['NAME'].reset_index()
```

```python
pop_2018['COUNTY']=='000'
```

## Find out how many county FIPS CODE
## Logic is in each state, COUNTY FIPS CODE with non-zero considers as county and place code is 00000, CONSUB is 00000

```python

county_df = pop_2018[(pop_2018['COUNTY']!='000')&(pop_2018['PLACE']=='00000')&(pop_2018['COUSUB']=='00000')]
county_df.reset_index(drop=True)
```

```python
print("There are {} County FIPS CODE in 2018 population data.".format(county_df.shape[0]))
```

## Which meets excel pivot table number.

```python
# After we have out how many county code, I want to find out how many counties are in each state and show top 5 states

county_code = county_df.groupby(['STATE', 'STNAME']).agg(['count'])['NAME'].\
sort_values(by='count', ascending=False).reset_index()

county_code.head()
```

## I checked google also and number are matched.

## find out how many PLACE FIPS CODE
## Logic is in each state, PLACE FIPS CODE with non-zero considers as city and COUNTY code is 00000,CONCIT is 00000

```python

city_df = pop_2018[(pop_2018['PLACE']!='00000')&(pop_2018['COUNTY']=='000')&(pop_2018['CONCIT']=='00000')]
city_df.reset_index(drop=True)
```

```python
print("There are {} PLACE FIPS CODE in 2018 population data.".format(city_df.shape[0]))
```

```python
city_code = city_df.groupby(['STATE', 'STNAME']).agg(['count'])['NAME'].\
sort_values(by='count', ascending=False).reset_index()

city_code.head()
```

```python
# Combine county and city dataset
join_df = pd.merge(county_code, city_code, on=['STATE', 'STNAME'],suffixes=('_County', '_Place'))
join_df
```

```python
join_df['Total FIPS CODE'] = join_df['count_County']+join_df['count_Place']
```

```python
final_df = join_df.sort_values(by='Total FIPS CODE', ascending=False).reset_index(drop=True)
```

```python
final_df
```

```python
final_df.to_csv("joindf.csv",index=False)
```

# Change in FIPS CODE

How many FIPS codes have been changed. (especially in place code level)
For change, I downloaded the data from Geographic Boundary Change Notes in Census.


https://www.census.gov/programs-surveys/geography/technical-documentation/boundary-change-notes.html
I chose the United States to get all the dataset. 
https://www2.census.gov/geo/docs/reference/bndrychange/united-states.txt




Also in this file, before pulling up the file, I want to choose coltype because FIPS CODE is numerical code, but many programs recoginized as numerical code, so it deletes leading zero automatically.

```python
# Preselect column type as object
coltype = {'State':object,
           'Code':object}
```

```python
changes_fips_code = pd.read_csv("Boundary_change.csv", dtype= coltype)
```

```python
changes_fips_code
```

There are 8,614 cases of change from 1870 - 2010

Then, How many changes happened in a year?

```python
changes_fips_code.groupby('Effective_Year').agg('count')['Code']
```

```python
# Check what will be the biggest cause of change
# I will groupby Type of Change
reason_df = changes_fips_code.groupby('Type of Change')['Code'].agg('count').reset_index().sort_values(by='Code', ascending=False)
```

```python
#Top 5 Reason
reason_df.rename(columns={"Code": "Number of Change"}, inplace=True)
reason_df.head().reset_index(drop=True)
```

```python
# Let's check changes after year 1999 because that is why most changes are occurred.

changes_after_1999= changes_fips_code[changes_fips_code['Effective_Year']> 1998]
```

```python
reason_df_1999 = changes_after_1999.groupby('Type of Change')['Code'].agg('count').reset_index().sort_values(by='Code', ascending=False)
```

```python
reason_df_1999.rename(columns={"Code": "Number of Change"}, inplace=True)
reason_df_1999.head().reset_index(drop=True)
```
